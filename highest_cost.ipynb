```python
# imports
import pandas as pd
import numpy as np
```

```python
# importing csv's
claims = pd.read_csv('claims.csv')
members = pd.read_csv('members.csv')
```

```python
# Adding a claim count column to track the amount of claims for groupby purposes
claims['claim_count'] = 1
```

```python
# Creating a new df grouping cpt codes
cpt_code_claims = claims.groupby('cpt_code').sum()

# Parsing down to only the columns needed for this calculation
cpt_code_claims = cpt_code_claims[['paid_amount', 'claim_count']]

# Adding a new cloumn tracking the average paid per claim
cpt_code_claims['average_per_claim'] = (cpt_code_claims['paid_amount']/cpt_code_claims['claim_count']).round(2)

# Sorting values by average_per_claim and returning the top 10 highest
cpt_code_claims.sort_values('average_per_claim', ascending=False).head(10)
```

```python
# Creating a new df grouping icd codes
icd_code_claims = claims.groupby('icd_code').sum()

# Parsing down to only the columns needed for this calculation
icd_code_claims = icd_code_claims[['paid_amount', 'claim_count']]

# Adding a new column tracking the average paid per claim
icd_code_claims['average_per_claim'] = (icd_code_claims['paid_amount']/icd_code_claims['claim_count']).round(2)

# Sorting values by average_per_claim and returning the top 10 
icd_code_claims.sort_values('average_per_claim', ascending=False).head(10)
```
