```python
# imports
import pandas as pd
import numpy as np
```

```python
# importing csv's
claims = pd.read_csv('claims.csv')
members = pd.read_csv('members.csv')
```

```python
# Billed total calculation
total_billed = claims['billed_amount'].sum().round(2)

# Paid total calculation
total_paid = claims['paid_amount'].sum().round(2)

# Difference between billed and paid amounts
total_difference = total_billed - total_paid

# Total number of claims
num_claims = claims['claim_id'].count()

# Average Unpaid Claim Cost calculation
avg_unpaid = (total_difference / num_claims).round(2)
```

```python
# Sorting claims by paid_amount
claims.sort_values('paid_amount', ascending=False)
```

```python
# Create new df with only the columns needed
cpt_code = claims[['cpt_code', 'paid_amount']].copy()

# Grouping and sorting cpt_code by highest paif amount
highest_cpt_code = cpt_code.groupby('cpt_code').sum().sort_values('paid_amount', ascending=False)
```
