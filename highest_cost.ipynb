```python
total_billed = claims['billed_amount'].sum().round(2)
total_paid = claims['paid_amount'].sum().round(2)
total_difference = total_billed - total_paid
print(f'The total billed amount is ${total_billed}.')
print(f'The total paid amount is ${total_paid}.')
print(f'The total difference is ${total_difference.round(2)}.')
```
